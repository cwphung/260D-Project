```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
partitions = np.array([1, 2, 3, 4, 8, 16, 24, 32])
wg_acc_shuffle = np.array([26.45, 43.05, 27, 26.2, 26.95, 20.05, 16.84, 0])
wg_acc = np.array([26.45, 15.87, 18.64, 15.87, 20.86, 14.11, 18.38, 0])
avg_acc_shuffle = np.array([99.37, 98.8, 99.22, 99.48, 99.09, 99.27, 99.43, 99.39])
```

```python
plt.plot(partitions, wg_acc_shuffle, label='Shuffled')
plt.plot(partitions, wg_acc, label='Unshuffled')
plt.xlabel('Num Partitions')
plt.ylabel('Worst-group Accuracy')
plt.title('Worst-group Accuracy vs. Num Partitions')
plt.xlim([1, 32])
plt.ylim([0, 50])
plt.legend()
```
